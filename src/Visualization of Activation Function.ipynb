```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x * np.sin(5 * x) * np.exp(-0.5 * np.square(x))

# Generate x values
x = np.linspace(-3, 3, 500) 

# Compute y values
y = f(x)

# Plot the function
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$x \sin(5x) e^{-0.5x^2}$', color='blue')
plt.title('Plot of $x \sin(5x) e^{-0.5x^2}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```
